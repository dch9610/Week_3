## 간단한 사이트에서 수집

```python
# 1. 모듈 가져오기
from bs4 import BeautifulSoup
from urllib.request import urlopen
```

```python
# 2. 타겟 사이트 접속
target_site ='https://finance.naver.com/marketindex/exchangeList.nhn'
res =urlopen(target_site)
```

```python
if res.getcode() != 200:
    print('사이트에 문제가 있다.')
```

```python
# 3. 응답받은 데이터 => html 데이터(반정형 데이터)
# html 데이터를 파싱하는 모듈 -> 원하는 데이터를 추출 -> bs4에서 해결
# 홈페이지에서 parser를 검색하여 html5lib 설치
# soup = BeautifulSoup(res,파서의 종류 )
soup = BeautifulSoup(res, 'html5lib' )
soup
```

```python
# 4. 통화명 추출 : 1) 특정 2) 추출
# css selector : .tit
# 우클릭 후 copy에 가서 selector copy를 선택
# body > div > table > tbody > tr:nth-child(1) > td.tit
```

```python
# soup.select() => 모두 다 찾아라 -> 복수 -> 리스트
# soup.select_one () => 한개만 찾아라 -> 단수 -> 값
tmp = soup.select('.tit')

# 파싱한 결과는 DOM tree가 메모리에 올라가게 되고 
# <td></td> 이런것들이 전부다 객체로 표현되기 떄문에
# 그 이하를 접근할때는 .을 통해서 접근한다
len(tmp), type(tmp), tmp[0].a
```

```python
print(tmp[0].a.string.strip())
```

```python
# 전체 통화 추출
for td in soup.select('.tit'):
    print(td.a.string.strip())
```

```python
# 각 국가별 매매기준율 추출하시오
# body > div > table > tbody > tr:nth-child(1) > td.sale
sale = soup.select('.sale')
sale
```

```python
# 전체 매매기준율 추출
for sale in soup.select('.sale'):
    print(sale.string.strip().replace(',', ''))
type(sale)
```

## 요구사항

- 통화명, 매매기준율, 사실때, 보내실때, 통화 코드 추출
- 이런 데이터를 추출하기 위해서 사이트의 html의 구조를 꼼꼼하게 체크, 
    - 개발자도구 > Elements 파트에서 시뮬레이션을 통해서 특정한 것이 정확한지 체킹
- 반복적으로 등장하는 데이터 (rows, 리스트형태(tr, li 태그로 나열된 형태))
    - row 데이터 단위로 수집을 수행해야, 문제 없이 수집이 가능!
    - [{},{},{}, ...] 구조로 손쉽게 구성

```python
# 통화정보를 가진 tr 44개를 수집하시오
# body > div > table > tbody > tr:nth-child(1)
tmp = soup.select('tbody>tr')
len(tmp)
```

```python
for tr in tmp:
    # name 출력
    print(tr.select_one('.tit').a.string.strip())
    
    # code 출력
    # 속성값 불러오는 방법
    # print(tr.select_one('.tit').a['href'])
    print(tr.select_one('.tit').a.get('href')[-6:-3])
    
    # buy_std 출력
    print(tr.select_one('.sale').string.replace(',',''))
    
    # buy_cash출력
    print(tr.select_one('td:nth-child(3)').string.replace(',',''))
    
    # send 출력
    # nth-child => nth-of-type으로 변경해서 사용가능
    print(tr.select_one('td:nth-of-type(5)').string.replace(',',''))    
    break
```

```python
result = list() # 나중에 추가되는 동적 스타일
for tr in tmp:
    dic={
        'name'     : tr.select_one(".tit").a.string.strip(),
        'code'     : tr.select_one(".tit").a.get("href")[-6:-3],
        'buy_std'  : tr.select_one(".sale").string.replace(',',''),
        'buy-cash' : tr.select_one("td:nth-child(3)").string.replace(',',''),
        'send'     : tr.select_one("td:nth-child(5)").string.replace(',','')
        }
    result.append(dic)
result
```

```python
# json 저장 (db 저장도 가능: RDBS, No-SQL)
import json
```

```python
# 저장 (save, dump)
with open('fx.json', 'w') as fp:
    json.dump (result,fp)
```

```python
# 로드
with open('fx.json', 'r') as fp:
    print(json.load (fp))
```

```python
# 3. 4개 파일을 다운로드 할수 있는 문자열 추출 리스트
# body > p:nth-child(10) > tt:nth-child(1) > a

# print(tr.select_one('.tit').a.get('href')[-6:-3]
tmp=soup.select('tt:nth-child(1)')

name = soup.select('tt')[:4]
len(soup.select('tt')[:4])
```

```python
# 4. 파일 다운로드
# 원격지에 있는 파일과, 로컬 상에 특정위치(저장한 파일)까지 스트림을 열고
# 읽어서 바로 기록 -> 파일 저장에 필요한 팝업이 생략, 원하는 위치에 바로 저장
import os, os.path
```

### 5. 압축 데이터 해제
- gz 파일
    - gzip 모듈을 활용


```python
import gzip
```

```python
for file in tqdm.notebook.tqdm(files):
    # 원본파일경로
    path = os.path.join(savedPath, file)
    # 대상 파일경로 => .gz를 제거 파일명을 생성
    rawFile_path = os.path.join(savedPath, file[:-3])
    # print(rawFile_path)
    
    # 압축파일 오픈 => 읽기 (용량이 크면 나눠서, 작으면 한번에) => 기록
    with gzip.open(path, "rb") as fg:
        # 읽기 (한번에) -> 압축이 해제된 내용으로 읽힌다.
        tmp = fg.read()
        # 쓰기 -> 파일오픈
        with open(rawFile_path,'wb') as f:
            # 기록
            f.write(tmp)
        
```
