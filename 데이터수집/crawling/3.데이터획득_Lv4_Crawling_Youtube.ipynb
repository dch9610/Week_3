```python
# 1. 모듈 가져오기
from selenium import webdriver as wd
import platform
import time
import urllib # 인코딩하기위해
```

```python
# 브라우저 띄우기
if platform.system() == 'Windows':
    driver = wd.Chrome('./chromedriver.exe')
else: # Darwin:
    driver = wd.Chrome('./chromedriver')
```

```python
# 3. 접속
keyword = urllib.parse.quote('손흥민')
target_site = f'https://www.youtube.com/results?search_query={keyword}&sp=EgIIAg%253D%253D'
target_site
```

```python
# 해당 사이트로 접속
driver.get( target_site )
```

```python
# 4. 더보기 액션을 작동시켜서, 1일치 데이터를 전부 로드한다
#    상위버전 -> 더보기 통신 프로토콜을 분석하여 직접 호출한다 (조심, 많은 검토 필요)
# 스크롤바를 밑으로 내리는 행위, 스크립트에서 스크롤 이동 처리
for n in range(30): # 10회 정도만 내리겠다 (설정)
    # 현재 화면에서 스크립트를 가동
    # 넉넉하게 크기(1000)를 잡음
    driver.execute_script('''
        window.scrollBy(0,1000)
    ''')
    time.sleep(2)
```

```python
# 5. 썸네일, 제목, 링크, 조회수, 업로드시간, 게시자, 영상요약, 댓글 ...
# 영상의 메타 데이터 수집
# 검색 결과는 ytd-video-renderer 태그를 모으면 된다
videos = driver.find_elements_by_tag_name('ytd-video-renderer')
len( videos )
```

```python
# 최신 2건만 진행
for video in videos[:2]:
    # 제목
    target = video.find_element_by_id('video-title')
    title  = target.get_attribute('title')
    # 썸네일
    thumbnail = video.find_element_by_id('img').get_attribute('src')
    # 링크
    video_url = target.get_attribute('href')
    print(title, thumbnail, video_url)
    
```

```python
video_metas = list()
for video in videos[:2]:
    target    = video.find_element_by_id('video-title')
    title     = target.get_attribute('title')
    thumbnail = video.find_element_by_id('img').get_attribute('src')
    video_url = target.get_attribute('href')
    video_metas.append({
        't':title,
        'n':thumbnail,
        'v':video_url
    })
```

```python
# 리뷰 로드
for page in video_metas[:1]:
    driver.get(page['v'])
    time.sleep(5)
    
    # 리뷰를 로드하기 위해서 컨텐츠를 스크롤한다.
    for n in range(10): 
        driver.execute_script('''
            window.scrollBy(0,1000)
        ''')
        time.sleep(2)
    # 리뷰 수집
    reviews = driver.find_elements_by_tag_name('ytd-comment-thread-renderer')
    for review in reviews:
        print('-'*100)
        # 해당 요소 밑에 있는 모든 텍스트를 가져오시오
        print(review.text)
        print('-'*100)
```
